```python
from numpy import arange
import numpy as np
import matplotlib.pyplot as plt
from math import cos,sin,pi
```

```python
def func(m,x,the):
    return (1/pi)*cos(m*the-x*sin(the))    
```

```python
#regla de simpson
def Jm(x,m):
    
    a=0
    b=pi
    fa=func(m,x,a)
    fb=func(m,x,b)
    N=1000
    h=(b-a)/N
    Si=0
    Sp=0
    for k in range ((N//2)+1):
        Si+=func(m,x,a+(2*k-1)*h)
        if k <(N//2): 
            Sp+=func(m,x,a+(2*k*h))
    
    return (h/3)*(fa+fb+4*Si+2*Sp)
    
```

```python
x=[]
J0=[]
J1=[]
J2=[]
for i in np.arange (0,21,0.1):
    x.append(i)
    J0.append(Jm(i,0))
    J1.append(Jm(i,1))
    J2.append(Jm(i,2))
```

```python
plt.plot(x,J0)
plt.plot(x,J1)
plt.plot(x,J2)
plt.show()
```

```python
l=500 #nm (longitud de onda)
k=(2*pi)/l #numero de onda

def int(r):
    return (Jm(k*r,1)/(k*r))**2

plt.imshow(int())
```
