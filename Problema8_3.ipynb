**Ecuaciones de lorenz**

\begin{equation}
\frac{dx}{dt}=\sigma (y-x)
\end{equation}

\begin{equation}
\frac{dy}{dt}=rx-y-xz
\end{equation}

\begin{equation}
\frac{dz}{dt}=xy-bz
\end{equation}


Estas ecuaciones fueron estudiadas por primera vez por Edward Lorenz en 1963, quien las derivó de un modelo simplificado de patrones climáticos. La razón de su fama es que fueron uno de los primeros ejemplos incontrovertibles del caos determinista, la aparición de un movimiento aparentemente aleatorio a pesar de que no hay aleatoriedad incorporada en las ecuaciones.

```
from math import sin
from numpy import array, arange
from pylab import plot, xlabel, ylabel, show, title
```

a) Escribimos un programa para resolver las ecuaciones de Lorenz para los valores σ = 10, r = 28 y b = 8/3
En el rango de t = 0 a t = 50 con condiciones iniciales (x, y, z) = (0, 1, 0)

```
def f(R,t):    # Definimos una funcion que resive como parametro un arreglo R que contiene los valores de las variables x,y,z, y al tiempo
    b = 8/3    
    r = 28
    s = 10
    x = R[0]
    y = R[1]
    z = R[2]
    fx = s*(y-x)
    fy = r*x-y-x*z
    fz = x*y-b*z
    return array( [fx,fy,fz,],float)   # la funcion retorna un arreglo con las derivadas de las variables x,y,z.
```

```
ti = 0
tf = 50
N = 1000
h = (tf-ti)/N
 
R= array([0.0,1.0,0.0], float)   # el arreglo R tiene en principio los valores valores inciales de x,y,z y posteriormente se va llenando mediante el ciclo for para pasarlos a la funcion.

xpoints = []
ypoints = []
zpoints = []

tpoints = arange(ti,tf,h)

for t in tpoints:       # mediante el ciclo for implementamos el metodo de Runge-kutta de 4 orden y rretornamos los valores para x,y.z
    xpoints.append(R[0])
    ypoints.append(R[1])
    zpoints.append(R[2])

    k1 = h*f(R,t)
    k2 = h*f(R+0.5*k1,t+0.5*h)
    k3 = h*f(R+0.5*k2,t+0.5*h)
    k4 = h*f(R+k3,t+h)
    R += (k1+2*k2+2*k3+k4)/6
```

**a)** graficamos a Y en función del tiempo. y podemos observar que tenemos una funcion caotica.

```
plot(tpoints,ypoints)
xlabel('t')
ylabel('y(t)')
show()
```

**b)** Ahora realizamos una gráfica de z contra x. y vemos la imagen del
famoso "atractor extraño" de las ecuaciones de Lorenz, una funcion en forma de mariposa nunca se repite, A diferencia de los atractores clásicos, los atractores extraños tienen estructura a todas las escalas. Un atractor es extraño si la dinámica en el atractor es caótica.

```
plot(xpoints,zpoints)
xlabel('x')
ylabel('z')
title('atractor extraño')
show()
```
